```python
import numpy as np
import pandas as pd
import csv
```

```python
anime = pd.read_csv("anime.csv")
reviews = pd.read_csv("rating.csv")
```

```python
all_anime=[]

for i in reviews.user_id.unique():
    single_user = reviews[reviews.user_id == i].values
    new = {}
    for j in single_user:
        new[j[1]] = j[2]
    #print(new)
    all_anime.append(new)

    
```

```python
all_anime.insert(0,[])
```

```python
with open("new_ratings.csv","a",newline='') as myfile:
    csv_write = csv.writer(myfile)
    for i in all_anime:
        csv_write.writerow([i])
```
